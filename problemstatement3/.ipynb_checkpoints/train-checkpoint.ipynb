```python
# Step 1

import pandas as pd
from IPython.display import display
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle 
```

```python
data = pd.read_csv("loan_approval_data.csv")
```

```python
data
```

```python
# data = data.dropna()
```

```python
# Step 2
display(data.sample(10))
display(data.shape)
display(data.describe())
```

```python
# Dependants is considered as categorical because realistically, there will only be a limited, finite number of dependents
categorical_features = ['gender', 'married', 'dependents', 'education', 'self_employed', 'property_area', 'loan_status']

# Check out all the unique values
for col in categorical_features:
    col_data = data[[col]]
    display(f"For col={col}, unique values = {pd.unique(col_data.values.reshape(-1))}")
```

```python
# Step 3

data.drop('loan_id', axis=1, inplace=True, errors='ignore')
data.drop_duplicates(inplace=True)
```

We will not ignore entries with "isNA" [which is the case for gender, married, dependents, self_employed, loanamount, credit_history, loan_amount_term) because it can be considered as signals for the loan approval process. 

It is generally expected that loan applicants provide all the above information, as they can be detrimental to their capability of paying back the loan. Maybe certain fields (like gender) can be dropped off to avoid any potentially negative bias.

```python
# Step 4
plt.rcParams['figure.figsize'] = [10,30]
fig, ax = plt.subplots(len(categorical_features), 1)

# Check the loan_status distribution for every categorical feature
for i, col in enumerate(categorical_features):
    sns.countplot(data, x=col, hue='loan_status', ax=ax[i])

fig.show()
```

```python
# Step 5

label_encoder = LabelEncoder()
for col in categorical_features:
    if col == "dependents":
        # In a way, it is already categorical encoded
        continue
    data[col] = label_encoder.fit_transform(data[col])
```

```python
display(data)
```

```python
# Step 6

X = data.drop('loan_status', axis = 1)
Y = data['loan_status']
```

```python
X
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=1)

display(X_train.shape, X_test.shape)
display(Y_train.shape, Y_test.shape)
```

```python
# scaler = StandardScaler()

# X_train = scaler.fit_transform(X_train.iloc[:,:])
# X_test = scaler.transform(X_test.iloc[:,:])
```

```python
## Step 7
model = RandomForestClassifier(
                               n_estimators=300, 
                               max_depth=7,
                                max_features=None
                              )
model.fit(X_train, Y_train)

print(f"Train Accuracy = {100*model.score(X_train, Y_train)}%")

print(f"Test Accuracy = {100*model.score(X_test, Y_test)}%")
```

```python
pickle.dump(model, open('backend/model.pkl','wb'))
```

```python
from sklearn.tree import plot_tree

for tree_in_forest in model.estimators_:
    plot_tree(tree_in_forest)
    break
```
